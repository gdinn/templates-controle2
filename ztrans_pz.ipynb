```python
from lcapy import delta, u
from lcapy.discretetime import n, z

from matplotlib.pyplot import figure, plot, grid, subplots, title, xlabel, ylabel
from numpy import linspace


#x = (0.5**n)*u(n);
x = delta(n) + 2 * delta(n - 2)

figure();
ax = x.plot();
ax.set_title("Função avaliada");

X = x(z)
figure();
ax = X.plot();
ax.set_title("Transformada Z");

t = linspace(0,1,1000+1);
Xt = [abs(item.sympy) for item in X(t)];
figure();
ax = plot(t,Xt);
title("Checagem da ROC");
xlabel("Z");
ylabel("X(Z)");
grid();

print("Transformada Z");
print(X)
```
